```python
import pandas as pd
```

```python
ds = pd.read_csv('Housing.csv')
```

```python
ds.head()
```

```python
ds.tail(5)
```

```python
ds.isnull().sum()
```

```python
ds.head()
```

```python
listpreprocess = ['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']
```

```python
ds[listpreprocess] = ds[listpreprocess].replace({'yes':1,'no':0})
```

```python
ds.head()
```

```python
ds['furnishingstatus'] = ds['furnishingstatus'].replace({'furnished':2,'semi-furnished':1,'unfurnished':0})
```

```python
ds.head()
```

here we can see the price and area are in the counts of thousands and lakhs/millions , it is better practice to scale down this value for better ML understanding , so we are going to standardize these two datas as we did with other datas

```python
from sklearn.preprocessing import StandardScaler as scale
```

```python
listpreprocess2 = ['area','price']
```

```python
ds[listpreprocess2] = scale().fit_transform(ds[listpreprocess2])
```

```python
ds.head()
```

```python
ds.info()
```

the data preprocessing is completed and now we are going to start the data visualization part


```python
datacorrelation = ds.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
plt.figure(figsize=(9,9))
sb.heatmap(datacorrelation,annot= True,cmap='Blues')
plt.show()
```

```python
ds.hist(figsize=(9,9),bins=10)
plt.show()
```

now data visualizaion part is completed now we have to start ML process . here we are going to use linear regression method


```python
x = ds.drop('price' , axis= 1)
y = ds['price']
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4,random_state=40)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR= LinearRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
y_predict = LR.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
LR_accuracy = r2_score(y_test,y_predict)
```
